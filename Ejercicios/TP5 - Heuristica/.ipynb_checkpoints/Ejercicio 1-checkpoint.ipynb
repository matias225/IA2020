# El problema del viajante

Dada una lista de ciudades y las distancias entre cada par de ciudades determinar la ruta más pequeña posible que visite cada ciudad y regrese a la ciudad de origen.

El problema en esta situación consiste en ser un problema de combinatoria donde el costo computacional de evaluar todas las combinaciones posibles aumenta mientras se incrementa la cantidad de ciudades.

Para obtener lo datos para la resolución y la heurística puede utilizarse la herramienta Google Maps.

# Primera parte

Resolver el problema utilizando un método de fuerza bruta para los siguientes destinos:

- San Rafael
- El sosneado
- Jaime Prats
- Gral. Alvear

Mostrar el costo de todas las rutas

Distancias:
    _ San Rafael - Sosneado: 188 km
    _ San Rafael - Jaime Prats: 70.1 km
    _ San Rafael - Gral. Alvear: 84.8 km
    
    _ Sosneado - San Rafel: 188 km
    _ Sosneado - Jaime Prats: 251 km
    _ Sosneado - Gral. Alvear: 266 km
    
    _ Jaime Prats - San Rafel: 70.1 km
    _ Jaime Prats - Sosneado: 251 km

    _ Gral Alvear - San Rafel: 84.7 km
    _ Gral Alvear - Jaime Prats: 14.9 km
    _ Gral Alvear - Sosneado: 266 km

```python
#a = San Rafael
#b = Jaime Prats
#c = Gral Alvear
#d = Sosneado

#lista = [1, 2, 3, 4, 1]
#lista = [1, 2, 4, 3, 1]
#lista = [1, 3, 2, 4, 1]
#lista = [1, 3, 4, 2, 1]
#lista = [1, 4, 2, 3, 1]
#lista = [1, 4, 3, 2, 1]

from itertools import permutations
import itertools

ciudades = [[0, 70.1, 84.8, 188],
            [70.1, 0, 14.9, 251],
            [84.7, 14.9, 0, 266],
            [188, 251, 266, 0]]

ciudades2 = ["San Rafael", "Jaime Prats", "Gral Alvear", "Sosneado"]

for item in itertools.permutations(ciudades2[1:]):
    costeruta = []
    aux = 0
    item = list(item)
    item.insert(0,"San Rafael")
    item.append("San Rafael")
    print(item)
    i=0
    for item2 in range(len(item)-1):
        print(ciudades2.index(item[i]))
        print(ciudades2.index(item[i+1]))
        print(item[i] + "-" + item[i+1])
        #print(ciudades[ciudades2.index(item[i])][ciudades2.index(item[i+1])])
        aux = aux + ciudades[ciudades2.index(item[i])][ciudades2.index(item[i+1])]
        i=i+1
    costeruta.append(aux)
    print(aux)
    
print("Coste ruta menor: ", min(costeruta) ,"km")
    
```

# Conclusion

El viaje con menor costo es ['San Rafael', 'Sosneado', 'Jaime Prats', 'Gral Alvear', 'San Rafael'] con un costo en km de 538.6.
